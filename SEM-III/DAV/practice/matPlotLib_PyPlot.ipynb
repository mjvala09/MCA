```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["one","two","three","four"]
y=[35,40,25,50]
z=[40,50,60,70]

p=np.arange(len(x))
p1=[i+0.2 for i in p]

c=["r","g","y","b"]
plt.bar(p,z,0.2,color="g")
plt.bar(p1,y,0.2,color="r")

plt.title("DEMO",fontsize=20) #autopct="%0.2f%%"
plt.xlabel("Sequences")
plt.ylabel("populations")

plt.xticks(p+0.2/2,rotation=20)
plt.legend("population")

plt.show()
```
