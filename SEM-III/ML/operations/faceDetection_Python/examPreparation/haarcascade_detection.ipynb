```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../img4.jpg')
```

```python
face_classifier = cv2.CascadeClassifier("../../haarcascade_files/haarcascade_frontalface_alt.xml")
face = face_classifier.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))

for x, y, w, h in face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),5)

plt.title("face detection",color="g",fontsize=100)
plt.imshow(img)
```

```python
eye_classifier = cv2.CascadeClassifier("../../haarcascade_files/haarcascade_eye.xml")
eye = eye_classifier.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(40,40))

for x, y, w, h in eye:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,220,0), 4)

plt.imshow(img)
plt.title("Eye Detection", color="r",fontsize=100)
plt.show()
```

```python
leftEerImg = cv2.imread("../left-ear.jpg")
rightEarImg = cv2.imread("../right-ear.jpg")
leftEerClassifier = cv2.CascadeClassifier("../../haarcascade_files/haarcascade_mcs_leftear.xml")
rightEerClassifier = cv2.CascadeClassifier("../../haarcascade_files/haarcascade_mcs_rightear.xml")
```

```python
leftEar = leftEerClassifier.detectMultiScale(leftEerImg, scaleFactor=1.1, minNeighbors=5, minSize=(40,40))

for x,y,w,h in leftEar:
    cv2.rectangle(leftEerImg, (x,y), (x+w,y+h), (0,250,0), 10)

plt.imshow(leftEerImg)
plt.title("Left Ear Detection", color="orange", size=100)
plt.show()
```

```python
rightEar = rightEerClassifier.detectMultiScale(rightEarImg, scaleFactor=1.1, minNeighbors=5, minSize=(40,40))

for x,y,w,h in rightEar:
    cv2.rectangle(rightEarImg, (x,y), (x+w,y+h), (0,250,0), 10)

plt.imshow(rightEarImg)
plt.title("Left Ear Detection", color="orange", size=100)
plt.show()
```
