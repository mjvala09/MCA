```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20)
y = x**2
```

```python
plt.plot(x, y, 'r^')
plt.show()
```

# Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['one','two','three','four']
numbers = [23,45,18,8]
ax.pie(numbers, labels = languages, autopct='%1.1f%%')
plt.show()
```
